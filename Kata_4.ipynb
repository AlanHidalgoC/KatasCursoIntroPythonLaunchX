# Ejercicio 1: Transformar cadenas
Hay varias operaciones que puedes realizar en las cadenas cuando las manipulamos. En este ejercicio, usar치s m칠todos de cadena para modificar el texto con hechos sobre la Luna y luego extraer치s informaci칩n para crear un breve resumen.



El texto con el que trabajar치s es el siguiente: 

````python
text = """Interesting facts about the Moon. The Moon is Earth's only satellite. There are several interesting facts about the Moon and how it affects life here on Earth. 
On average, the Moon moves 4cm away from the Earth every year. This yearly drift is not significant enough to cause immediate effects on Earth. The highest daylight temperature of the Moon is 127 C."""
````

Primero, divide el texto en cada oraci칩n para trabajar con su contenido:

````python
texto_dividido = text.split('.')
texto_dividido
````

Ahora, define algunas palabras clave para b칰squeda que te ayudar치n a determinar si una oraci칩n contiene un hecho.

````python
palabras_clave = ["average", "temperature", "distance"]
````

Cre un bucle para imprimir solo datos sobre la Luna que est칠n relacionados con las palabras clave definidas anteriormente:

````python
for datos in texto_dividido:
 for palabras_clave in palabras_clave:
  if palabras_clave in datos:
   print(datos)
   break
    
````

Finalmente, actualiza el bucle(ciclo) para cambiar _C_ a _Celsius_:

````python
for datos in texto_dividido:
 for palabras_clave in palabras_clave:
  if palabras_clave in datos:
   print(datos.replace('C', ' Celsius'))
   break
````
